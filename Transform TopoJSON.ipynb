```python
import json
import pycountry
```

```python
with open('sentinel/static/world-50m.json') as fh:
    contents = json.loads(fh.read())
```

```python
def transform(geo):
    try:
        country = pycountry.countries.get(numeric="{:03d}".format(geo['id']))
        geo['id'] = country.alpha3
        geo['properties'] = {'name': country.name}
    except KeyError:
        geo['properties'] = {'error': 1}
        
    return geo
        
contents['objects']['countries']['geometries'] = [transform(geo) for geo in contents['objects']['countries']['geometries']]
    
```

```python
kosovo = next(c for c in contents['objects']['countries']['geometries'] if c['arcs'] == [[-1092, 1174, -556]])
kosovo['id'] = "RS-KM"
kosovo['properties'] = {'name': "Kosovo"}
```

```python
with open('sentinel/static/world-50m-cnd.json', 'w') as fh:
    fh.write(json.dumps(contents))
```
